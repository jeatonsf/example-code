# Gaussian Kernel Smoother Plots on Regression Datasets

## Polynomials

```python
import matplotlib.pyplot as plt
import numpy as np

from gaussian_kernel_smoother import GaussianKernelSmoother


# Linear
x = np.random.uniform(-5, 5, size=512)
y = 2 * x + 1 + np.random.normal(0, 1, size=len(x))
y_pred = GaussianKernelSmoother().fit_predict(x, y)
plt.scatter(x, y, label="y=2x+1+noise")
idx = np.argsort(x)
plt.plot(x[idx], y_pred[idx], color="red", label="regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Quadratic
x = np.random.uniform(-5, 5, size=512)
y = 0.5 * x ** 2 + x - 1 + np.random.normal(0, 1, size=len(x))
y_pred = GaussianKernelSmoother().fit_predict(x, y)
plt.scatter(x, y, label="y=0.5x^2+x-1+noise")
idx = np.argsort(x)
plt.plot(x[idx], y_pred[idx], color="red", label="regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Cubic
x = np.random.uniform(-5, 5, size=512)
y = x ** 3 + x ** 2 + x + 1 + np.random.normal(0, 5, size=len(x))
y_pred = GaussianKernelSmoother().fit_predict(x, y)
plt.scatter(x, y, label="y=0.5x^2+x-1+noise")
idx = np.argsort(x)
plt.plot(x[idx], y_pred[idx], color="red", label="regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Nonlinear continuous functions

```python
import math


# Sine wave
x = np.random.uniform(-2 * math.pi, 2 * math.pi, size=1024)
y = np.sin(x) + np.random.normal(0, 1, size=len(x))
y_pred = GaussianKernelSmoother().fit_predict(x, y)
plt.scatter(x, y, label="y=sin(x)+noise")
idx = np.argsort(x)
plt.plot(x[idx], y_pred[idx], color="red", label="regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Damped sine wave
x = np.random.uniform(0, 5, size=1024)
y = np.exp(-x) * np.cos(2 * math.pi * x) + np.random.normal(0, 0.1, size=len(x))
y_pred = GaussianKernelSmoother().fit_predict(x, y)
plt.scatter(x, y, label="y=(e^-x)*cos(2pi*x)+noise")
idx = np.argsort(x)
plt.plot(x[idx], y_pred[idx], color="red", label="regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Noiseless increasing amplitude sine wave
x = np.random.uniform(0, 5, size=1024)
y = x * np.sin(2 * math.pi * x)
y_pred = GaussianKernelSmoother().fit_predict(x, y)
plt.scatter(x, y, label="y=x*(2pi*x)")
idx = np.argsort(x)
plt.plot(x[idx], y_pred[idx], color="red", label="regression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Sklearn's Diabetes Dataset

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_diabetes

from gaussian_kernel_smoother import GaussianKernelSmoother


X, y = load_diabetes(return_X_y=True)
feature_names = load_diabetes()["feature_names"]
for j in range(X.shape[1]):
    y_pred = GaussianKernelSmoother().fit_predict(X[:, j], y)
    plt.scatter(X[:, j], y, label="samples")
    idx = np.argsort(X[:, j])
    plt.plot(X[idx, j], y_pred[idx], color="red", label="regression")
    plt.xlabel(feature_names[j])
    plt.ylabel("target")
    plt.legend()
    plt.show()
```
